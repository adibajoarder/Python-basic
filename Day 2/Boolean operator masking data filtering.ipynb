```
import pandas as pd
```

```
data = {
    "name": ["John", "Anna", "Peter", "Linda"],
    "location": ["New York", "Paris", "Berlin", "London"],
    "age": [24, 13, 53, 33]
}
```

```
df = pd.DataFrame(data)
```

```
mask = df["age"] > 25
print(mask)
```

```
filtered_df = df[mask]
print(filtered_df)
```

```
mask = (df["age"] > 25) & (df["location"] == "London")
print(mask)
```

### Equality (`==`) Mask

```
mask_equal_location = df['location'] == 'London'
print("Mask for location == 'London':")
print(mask_equal_location)

mask_equal_age = df['age'] == 24
print("\nMask for age == 24:")
print(mask_equal_age)
```

### Inequality (`!=`) Mask

```
mask_not_equal_location = df['location'] != 'London'
print("Mask for location != 'London':")
print(mask_not_equal_location)

mask_not_equal_age = df['age'] != 24
print("\nMask for age != 24:")
print(mask_not_equal_age)
```

```
filtered_df_equal_location = df[mask_equal_location]
print("\nDataFrame filtered by location == 'London':")
display(filtered_df_equal_location)
```
